# Ejemplo alineamiento de pares de secuencias usando la libreria pairwise2 de biopython
#### Proteinas utilizadas P69905 , P68871

```python
from Bio import SwissProt
from Bio import pairwise2
from Bio import SeqIO
from Bio import ExPASy
```

```python
spraw1 = ExPASy.get_sprot_raw("P69905")
spinfo1 = SeqIO.read(spraw1,"swiss")
seq1 = str(spinfo1.seq)
```

```python
spraw2 = ExPASy.get_sprot_raw("P68871")
spinfo2 = SeqIO.read(spraw2,"swiss")
seq2 = str(spinfo2.seq)
```

```python
print(spinfo1.name)
print(spinfo2.name)
```

```python
ali = pairwise2.align.globalxx(seq1,seq2)
```

```python
ali
```

```python
from Bio.SubsMat.MatrixInfo import blosum62
```

```python
ali2 = pairwise2.align.localds(seq1, seq2, blosum62, -10, -1)
```
